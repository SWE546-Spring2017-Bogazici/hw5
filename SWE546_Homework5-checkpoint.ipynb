```python
# Define length with N
N = len(x)

def polynomialDegreeBasedPrediction(degree):
    
    A = np.hstack((np.power(x,i) for i in range(degree+1)))
    xtranspose = np.matrix(np.linspace(np.asscalar(x[0])-10,np.asscalar(x[-1])+10,30)).T
    AValues = np.hstack((np.power(xx,i) for i in range(degree+1)))

    # Least Squares Calculation 
    w_ls,E,rank,sigma = np.linalg.lstsq(A, y)
    
    # A(transpose)*A*w  
    formula = AValues*w_ls

    # Calculate prediction of year 2014 and compare it with actual of 2015
    x_calculations = np.matrix([2014,2015]).T-Start_Year
    A_calculations = np.hstack((np.power(x_calculations,i) for i in range(degree+1)))
    
    formula_calculate= A_calculations*w_ls
    
    ActualValue_2014 = float(cars[cars.Year == 2014].Car/1000000.)
    PredictedValue_2014 = formula_calculate[0,0]
    
    # Calculate accuracy for year 2014
    Year2014_Accuracy = 100 -(100 * abs(ActualValue_2014 - PredictedValue_2014) / ActualValue_2014)

    ActualValue_2015 = float(cars[cars.Year == 2015].Car/1000000.)
    PredictedValue_2015 = formula_calculate[1,0]
    
    # Calculate accuracy for year 2015
    Year2015_Accuracy = 100 - (100 * abs(ActualValue_2015 - PredictedValue_2015) / ActualValue_2015)
    
    AccuracyAVG = (Year2014_Accuracy + Year2015_Accuracy) / 2

    print('')
    print("\t DEGREE OF POLYNOMIAL IS " + str(degree))
    print('')
    
    plt.plot(xtranspose+Start_Year, formula, 'r',linewidth =2.0, color = 'violet')
    plt.plot(x+Start_Year, y, 'o-',linewidth =2.0, color = 'purple')

    # Plot the names of the axes y and x
    plt.xlabel('Years')
    plt.ylabel('Number of Cars')

    # Use matplotlib to call gca() to get the current axes
    plt.gca().set_ylim((-1,12))

    # Top and buttom values will be restricted with the help of set_ylim  and set_xlim
    plt.gca().set_xlim((1960,2020))

  
    print("Actual value for YEAR 2014:  \t" + str(ActualValue_2014))
    print("Predicted value for YEAR 2014: \t" + str(PredictedValue_2014))
    print("Actual value for YEAR 2015:    \t" + str(ActualValue_2015))
    print("Predicted value for YEAR 2014: \t" + str(PredictedValue_2015))
    print('')
    print("ACCURACY                       \t%" + str(AccuracyAVG))
    print('___________________________________________')
    print('')
    
    plt.show()
    
    return AccuracyAVG
```

```python
CalculateAccuracy = dict()

for degree in range(1, 11):
    CalculateAccuracy[degree] = polynomialDegreeBasedPrediction(degree)
```
