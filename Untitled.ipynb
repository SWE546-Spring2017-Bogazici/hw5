```python
%matplotlib inline

import scipy as sc
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt

df_arac = pd.read_csv(u'C:/Users/NedimAzgari/Desktop/SWE546/notes/data/arac.csv',sep=';')
df_arac[['Car','Year']]
```

```python
BaseYear = 1995
x = np.matrix(df_arac.Year[0:]).T-BaseYear
y = np.matrix(df_arac.Car[0:]).T
df_arac_reduced = df_arac

plt.plot(x+BaseYear, y, 'o-')
plt.xlabel('Yil')
plt.ylabel('Araba (Millions)')

plt.show()
```

```python
for i in range (13):
    degree=i
    A = np.hstack((np.power(x,deg) for deg in range(degree+1)))
    #print(A)
    w_ls,E,rank,sigma = np.linalg.lstsq(A, y)
    f = A*w_ls
    e = y-f
    eps = 0
    for l in range (len(e)): 
        eps = eps + e[l]**2
    eps = plt.sqrt(eps)
    plt.plot(x+BaseYear, y, 'o-')
    plt.plot(x+BaseYear, f, 'r')
    plt.xlabel('Years')
    plt.ylabel('Number of Cars')


    plt.show()
    print('Total error is ' , eps , ' for degree ' , i )
```
